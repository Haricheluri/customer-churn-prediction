# Phase 2: Data Cleaning

```python
import pandas as pd

churn = pd.read_csv("../data/raw/customer_churn.csv")
customers = pd.read_csv("../data/raw/customers.csv")
engagement = pd.read_csv("../data/raw/customer_engagement_metrics.csv")
rfm = pd.read_csv("../data/raw/customer_rfm.csv")
orders = pd.read_csv("../data/raw/orders.csv")
billing = pd.read_csv("../data/raw/subscription_billing.csv")
tickets = pd.read_csv("../data/raw/support_tickets.csv")
campaign = pd.read_csv("../data/raw/campaign_responses.csv")
pd.set_option('display.max_columns',None)
```

```python
datasets = {
    "customers": customers,
    "churn": churn,
    "engagement": engagement,
    "rfm": rfm,
    "orders": orders,
    "billing": billing,
    "tickets": tickets,
    "campaign": campaign
}
```

### country null value treatment

```python
customers[customers['country'].isna()].head()
```

```python
customers['state'].nunique()
```

based on states we treat the null values

```python
customers["country"] = customers["country"].fillna(
    customers.groupby("state")["country"]
             .transform(lambda x: x.mode()[0])
)
```

### orders null value treatment

```python
orders[orders['shipping_country'].isna()].head()
```

based on shipping state we treat null values for shipping country

```python
orders["shipping_country"] = orders["shipping_country"].fillna(
    orders.groupby("shipping_state")["shipping_country"]
          .transform(lambda x: x.mode()[0])
)
```

```python
orders[orders['promo_code'].isna()].head()
```

```python
orders["promo_code"] = orders["promo_code"].fillna(
    "No Promo"
)
```
