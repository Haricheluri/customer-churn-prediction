# Phase 4: Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option("display.max_columns", None)

df = pd.read_csv(
    "../data/processed/final_customer_data.csv"
)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="churned",
    data=df
)

plt.title("Customer Churn Distribution")

plt.show()
```

```python
important_num = [
    "tenure_months_x",
    "monthly_spend",
    "total_spend",
    "satisfaction_score",
    "nps_score_x",
    "predicted_churn_prob",
    "login_frequency_30d",
    "avg_session_duration_min",
    "recency_days",
    "frequency",
    "monetary",
    "total_orders",
    "total_revenue",
    "total_tickets"
]
```

```python
for col in important_num:
    plt.figure(figsize=(6,4))

    sns.histplot(
        df[col],
        kde=True
    )
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in important_num:
    plt.figure(figsize=(6,4))
    sns.boxplot(
        x=df[col]
    )
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
important_cat = [
    "gender",
    "marital_status",
    "segment",
    "contract_type_x",
    "payment_method_x",
    "engagement_tier",
    "account_status",
    "preferred_channel",
    "churn_reason"
]
```

```python
for col in important_cat:
    plt.figure(figsize=(8,4))
    sns.countplot(
        y=df[col],
        order=df[col].value_counts().index
    )
    plt.title(col)
    plt.show()
```

```python
churn_num = [
    "monthly_spend",
    "tenure_months_x",
    "satisfaction_score",
    "nps_score_x",
    "total_spend",
    "total_orders",
    "total_tickets",
    "frequency",
    "monetary"
]
```

```python
for col in important_num:
    plt.figure(figsize=(6,4))
    sns.boxplot(
        x="churned",
        y=col,
        data=df
    )
    plt.title(f"{col} vs Churn")
    plt.show()
```

```python
for col in important_cat:
    plt.figure(figsize=(8,4))

    sns.countplot(
        data=df,
        x=col,
        hue="churned"
    )
    plt.xticks(rotation=45)
    plt.title(f"{col} vs Churn")
    plt.show()
```

```python
corr_cols = [
    "monthly_spend",
    "total_spend",
    "satisfaction_score",
    "nps_score_x",
    "frequency",
    "monetary",
    "total_orders",
    "total_tickets",
    "churned"
]

plt.figure(figsize=(10,8))

sns.heatmap(
    df[corr_cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

## EDA Insights

1. Customer tenure has a strong relationship with churn. Customers with lower tenure are more likely to churn, while long-term customers tend to remain loyal.

2. Customers with lower monthly spending and total spending show a higher tendency to churn. High-value customers generally have better retention rates.

3. Satisfaction score significantly impacts churn. Customers with low satisfaction scores are more likely to leave the company.

4. NPS (Net Promoter Score) has a negative relationship with churn. Customers with higher NPS scores show stronger loyalty.

5. Predicted churn probability aligns well with the target variable, indicating that historical customer behavior is useful for identifying churn risk.

6. Customer engagement metrics such as login frequency and session duration indicate that highly engaged users are less likely to churn.

7. RFM metrics play an important role in customer retention. Customers with higher purchase frequency and monetary value tend to remain active.

8. Customers with a larger number of support tickets generally exhibit higher churn rates, indicating dissatisfaction or unresolved issues.

9. Contract type influences customer retention. Customers with longer contracts generally show lower churn rates.

10. Payment methods and engagement tiers show varying churn behaviors, indicating that customer interaction patterns affect retention.

11. High-value customers with greater total revenue and order counts are less likely to churn.

12. Correlation analysis shows that satisfaction score, NPS score, frequency, and monetary value have meaningful relationships with customer churn.

13. Several numerical features contain outliers, especially spending and revenue-related variables. These outliers represent genuine high-value customers and were retained.

14. Customer engagement, satisfaction, spending behavior, and support interactions are the most influential factors affecting churn.

15. The EDA indicates that customer churn is driven by a combination of behavioral, transactional, engagement, and satisfaction-related factors.

## Final Business Insights

• Low customer satisfaction increases churn risk.

• Low engagement customers are more likely to leave.

• Customers with more complaints and support tickets exhibit higher churn.

• Long-term customers demonstrate greater loyalty.

• High-spending customers have better retention rates.

• RFM metrics effectively identify valuable customers.

• Customer engagement significantly impacts retention.

• Contract duration influences churn behavior.

• Support quality plays a major role in customer retention.

• Customer satisfaction and NPS are among the strongest indicators of churn.
